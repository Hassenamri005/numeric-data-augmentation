# Data Augmentation

# So we will augment our data by adding 1 to the second column "Age" values five iterations

```python
# read
with open('test.csv', 'r') as f, open(r'test.csv', 'a') as myfile:
    f_csv = csv.reader(f)
    #Pass the header that contains "id" and "Age"
    header = next(f_csv)
    writer = csv.writer(myfile)
    for row in f_csv:
        print(row[1])
        for i in range(6):
          x = int(row[1])+i
          fields = [row[0], str(x)]
          print("Field : "+fields[0]+" "+fields[1])
          writer.writerow(fields)
```
